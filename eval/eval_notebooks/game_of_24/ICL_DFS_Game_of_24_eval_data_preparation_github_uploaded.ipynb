### Eliminate Puzzles included in the ICL

```python
puzzles_to_eliminate = ["1 1 9 9", "1 1 4 9", "4 6 8 11", "2 9 10 12"]
```

```python
game_24_data['Puzzles'].unique()
```

```python
game_24_data[game_24_data['Puzzles'] == "2 10 10 11"]
```

```python
for puzzle in puzzles_to_eliminate:
    game_24_data = game_24_data[game_24_data['Puzzles'] != puzzle]
```

```python
game_24_data.head()
```

### Randomly shuffle the puzzle

```python
import os, random
import numpy as np

def seed_everything(seed):
    os.environ['PYTHONHASHSEED'] = str(seed)
    random.seed(seed)
    np.random.seed(seed)

seed_everything(42)
```

```python
# shuffle game_24_data and pick any 30

sampled_df = game_24_data.sample(n=30, random_state=42).reset_index(drop=True)
```

```python
sampled_df.to_csv("eval_set_24.csv", index = False)
```
